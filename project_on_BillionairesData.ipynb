## Preparing Linear Regression model on Demographic Information y='finalWorth' x='gender', 'country', 'residenceStateRegion'  

```python
# turning categorical to dummy variable
```

```python
bdc_encoded = pd.get_dummies(bdc, columns=['gender', 'residenceStateRegion','industries','selfMade','status','category','source','country',
                                          'organization','title'])
```

```python
bdc_encoded.columns
```

```python
demographic_features =["age"]+["cpi_country"]+["cpi_change_country"]+[col for col in bdc_encoded.columns if col.startswith('gender_') or col.startswith('residenceStateRegion_')
                              or col.startswith('industries_')  or col.startswith('selfMade_')  or col.startswith('status_')
                                 or col.startswith('category_')or col.startswith('source_')or col.startswith('country_')
                              or col.startswith('organization_')or col.startswith('title_') ]
                                                                                                                                               
target = 'finalWorth'
```

```python
# now i will split the data 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = bdc_encoded[demographic_features]
y = bdc_encoded[target]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.28, random_state=42)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
model = LinearRegression()
model.fit(x_train, y_train)


y_pred = model.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(mse)
print(r2)
```

# Demographic Analysis

# Top Billionaires

```python
# Analyze the distribution of wealth among the top billionaires.
bdc.groupby(["personName","age"])[["finalWorth"]].sum().sort_values(by="finalWorth",ascending=False).head(10).plot(kind="bar",color="lightgreen",figsize=[10,5],label=True,title="Top 10 billionaires in the world")
```

```python
# Compare the net worth of individuals from different industries
bdc.groupby(["industries"])[["finalWorth"]].sum().plot(kind="bar",figsize=[10,5],color='lightgreen')
```

```python
##Country Wise Net worth of billionares

bdc.groupby(["country"])[["finalWorth"]].sum().sort_values(by="finalWorth",ascending=False).head(5)
```

```python
# Examine the dominant industries among the top billionaires.
bdc.groupby(["industries","personName"])[["finalWorth"]].sum().sort_values(by="finalWorth",ascending=False).head(10)
```

```python
# From which country and which industry Final worth of top billionare is high

bdc.groupby(["industries","country"])[["finalWorth"]].sum().sort_values(by="finalWorth",ascending=False).head(20).plot(kind="bar",figsize=[10,5],color='lightgreen')
```
